```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.linear_model import Perceptron

#Gera dados linearmente separáveis (2 caracterísitcas, 2 classes)
X, y = make_blobs(n_samples=100, centers=2, random_state=42, cluster_std=1.2)

#Iniciando e treinando o perceptron
#max_iter é o número máximo de vezes que ele passará pelos dados
modelo_perceptron = Perceptron(max_iter=1000, tol=1e-3, random_state=42)
modelo_perceptron.fit(X, y)

#Visualização gráfica:
#Criando uma malha (grid) para pintar o fundo conforme a predição
x_min, x_max = X[:,0].min()-1, X[:,0].max() +1
y_min, y_max= X[:,1].min()-1, X[:,1].max()+1

xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),np.arange(y_min, y_max, 0.02))

#Prevendo a classe para cada ponto da malha
Z = modelo_perceptron.predict(np.c_[xx.ravel(),yy.ravel()])
Z = Z.reshape(xx.shape)

#Plotando
plt.figure(figsize=(8,6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap="coolwarm")
plt.scatter(X[:,0], X[:,1], c=y, edgecolor='k', cmap='coolwarm')
plt.title('Fronteira de decisão do perceptron')
plt.xlabel('Característica 1 ')
plt.ylabel('Característica 2')
plt.show()
```
