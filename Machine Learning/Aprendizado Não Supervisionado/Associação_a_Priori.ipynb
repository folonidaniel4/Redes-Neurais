```python
#Carrega Bibliotecas
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
#Exemplo de Base de Dados
dataset = [['cerveja', 'pão', 'queijo', 'batata'],
    ['leite', 'pão', 'manteiga'],
    ['cerveja', 'pão', 'frutas'],
    ['leite', 'frutas', 'manteiga'],
    ['leite', 'pão', 'frutas', 'manteiga'],
    ['cerveja', 'pão', 'frutas'],
    ['pão', 'manteiga'],
    ['leite', 'pão', 'frutas', 'cerveja'],
    ['cerveja', 'pão', 'queijo', 'batata'],
    ['leite', 'pão', 'manteiga'],
    ['cerveja', 'pão', 'frutas'],
    ['leite', 'frutas', 'manteiga'],
    ['leite', 'pão', 'frutas', 'manteiga'],
    ['cerveja', 'pão', 'frutas'],
    ['pão', 'manteiga'],
    ['leite', 'pão', 'frutas', 'cerveja'],
    ['cerveja', 'pão', 'queijo', 'batata'],
    ['leite', 'pão', 'manteiga'],
    ['cerveja', 'pão', 'frutas'],
    ['leite', 'frutas', 'manteiga'],
    ['leite', 'pão', 'frutas', 'manteiga'],
    ['cerveja', 'pão', 'frutas'],
    ['pão', 'manteiga'],
    ['leite', 'pão', 'frutas', 'cerveja'],
    ['cerveja', 'pão', 'queijo', 'batata'],
    ['leite', 'pão', 'manteiga'],
    ['cerveja', 'pão', 'frutas'],
    ['leite', 'frutas', 'manteiga'],
    ['leite', 'pão', 'frutas', 'manteiga'],
    ['cerveja', 'pão', 'frutas'],
    ['pão', 'manteiga'],
    ['leite', 'pão', 'frutas', 'cerveja'],]
```

```python
#Transforma os dados em um formato apropriado para o algoritmo Apriori
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
#Encontrando os itens usando o Algoritmo a priori
frequent_itemsets = apriori(df, min_support=0.2, use_colnames=True)
```

```python
#Gerar as regras de associação
regras_associacao = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.5)
```

```python
#Imprime as regras de associação
print(regras_associacao)
```
