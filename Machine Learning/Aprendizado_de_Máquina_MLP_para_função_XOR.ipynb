```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
```

```python
#Dados para treinamento XOR
X_treino = np.array([[0,0],[0,1],[1,0],[1,1]])
y_xor = np.array([0,1,1,0])

#Criando um classificador MLP
mlp = MLPClassifier(hidden_layer_sizes=(2,1),activation='logistic',solver='lbfgs',learning_rate_init=0.1,max_iter=1000,random_state=42)

#Treinando o classificador MLP
mlp.fit(X_treino, y_xor)
```

```python
#Previsões para o conjunto de treinamento
prevision_xor = mlp.predict(X_treino)
print('Previsões para o XOR ',prevision_xor)
```

```python
#Plotando os dados de treinamento e as fronteiras de decisão
plt.figure(figsize=(8,6))
plt.scatter(X_treino[:,0], X_treino[:,1],c=y_xor, cmap='viridis', edgecolors='k', label='Treinamento')
plt.scatter(X_treino[:,0], X_treino[:,1], c=prevision_xor, cmap='viridis', marker='s', edgecolors='k', label='Previsões')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend(title='Classes')
plt.title('Dados de Treinamento e Fronteiras de decisões')
plt.show()
```
